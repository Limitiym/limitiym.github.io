```python
import numpy as np
import tensorflow as tf
%matplotlib inline
```

```python
train_dataset = tf.keras.preprocessing.image_dataset_from_directory("/content/Training", subset="training", seed=43, validation_split=0.1, batch_size=256, image_size=(100, 100))
validation_dataset = tf.keras.preprocessing.image_dataset_from_directory("/content/Training", subset="validation", seed=43, validation_split=0.1, batch_size=256, image_size=(100, 100))
```

```python
class_names = train_dataset.class_names
class_names
```

```python
import matplotlib.pyplot as plt
for images, lables in train_dataset.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[lables[i]])
    plt.axis("off")
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)
```

```python
from tensorflow import keras
import tensorflow as tf

AUTOTUNE = tf.data.experimental.AUTOTUNE
test_dataset = tf.keras.preprocessing.image_dataset_from_directory("/content/Validation", batch_size=256, image_size=(100, 100))
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)

model = keras.Sequential([keras.layers.Flatten(input_shape=(100, 100, 3)),
                          keras.layers.Dense(128, activation="relu"),
                          keras.layers.Dense(10, activation="softmax")
                          ])
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
history = model.fit(train_dataset, validation_data=validation_dataset, epochs=5, verbose=2)
score = model.evaluate(test_dataset, verbose=1)
print("Нейронка отрабатывает на столько процентов: ", round(score[1] * 100, 4))
```
